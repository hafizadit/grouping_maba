```python
import pandas as pd
import numpy as np
import os
import random
```

# Data Preparation

```python
fakultas = []

for i in os.listdir("data_csv"):
    fakultas.append(i[:-4])
```

```python
container = []
# Create df

df_FEB = pd.read_csv("data_csv/FEB.csv")
df_FH = pd.read_csv("data_csv/FH.csv")
df_FIB = pd.read_csv("data_csv/FIB.csv")
df_FISIP = pd.read_csv("data_csv/FISIP.csv")
df_FK = pd.read_csv("data_csv/FK.csv")
df_FKIP = pd.read_csv("data_csv/FKIP.csv")
df_FKOR = pd.read_csv("data_csv/FKOR.csv")
df_FMIPA = pd.read_csv("data_csv/FMIPA.csv")
df_FP = pd.read_csv("data_csv/FP.csv")
df_FSRD = pd.read_csv("data_csv/FSRD.csv")
df_FT = pd.read_csv("data_csv/FT.csv")
df_SV = pd.read_csv("data_csv/SV.csv")

container.extend([df_FEB, df_FH, df_FIB, df_FISIP, df_FK, df_FKIP, df_FKOR, df_FMIPA, df_FP, df_FSRD, df_FT, df_SV])
```

```python
# Add fakutlas COlumn

for data, fak in zip(container, fakultas):
    data["Fakultas"] = fak
```

#### Clear duplicate Value

```python
for i in container:
    try:
        i.drop_duplicates(subset ="Nama Lengkap",
                         keep='last', inplace = True)
    except:
        pass
```

```python
# Fix df_FH agar columns-nya sama

df_FH.drop(['Fakultas'], axis = 1, inplace=True)
```

```python
df_FH["Program Studi"] = "Hukum"
df_FH["Fakultas"] = "FH"
```

#### Convert to Dictionary

```python
data_dict = {}

for i in fakultas:
    data_dict[i] = []
```

```python
data_dict
```

```python
data_list = [i.values.tolist() for i in container]
```

```python
for i in data_list:
    print(i[:1])
    print("")
```

```python
for per_fak in data_list:
    for i in per_fak:
        data_dict[i[4]].append(i)
```

```python
# shuffle the data

for i in data_dict.keys():
    data_dict[i] = random.sample(data_dict[i], len(data_dict[i]))
```

# Algoritma

```python
data_dict["FEB"][0:4]
```

```python
kelompok_dict = {}

for i in range(1,300):
    kelompok_dict[i] = []
    check_temp = []
    iteration = 0
    while len(kelompok_dict[i]) < 38:
        for x in fakultas:
            if len(kelompok_dict[i]) == 38:
                break
            try:
                if x == "FH" or x == "FKOR" or x == "FSRD":
                    kelompok_dict[i].append(data_dict[x][0])
                    data_dict[x].pop(0)
                    if len(kelompok_dict[i]) == 38:
                        break
                elif x == "FEB" or x == "FMIPA":
                    for _ in range(3):
                        kelompok_dict[i].append(data_dict[x][0])
                        data_dict[x].pop(0)
                        if len(kelompok_dict[i]) == 38:
                            break
                elif x == "FIB" or x == "FISIP" or x == "FK":
                    for _ in range(2):
                        kelompok_dict[i].append(data_dict[x][0])
                        data_dict[x].pop(0)
                        if len(kelompok_dict[i]) == 38:
                            break
                elif x == "FKIP":
                    for _ in range(10):
                        kelompok_dict[i].append(data_dict[x][0])
                        data_dict[x].pop(0)
                        if len(kelompok_dict[i]) == 38:
                            break
                elif x == "FP" or x == "FT":
                    for _ in range(4):
                        kelompok_dict[i].append(data_dict[x][0])
                        data_dict[x].pop(0)
                        if len(kelompok_dict[i]) == 38:
                            break
                elif x == "SV":
                    for _ in range(5):
                        kelompok_dict[i].append(data_dict[x][0])
                        data_dict[x].pop(0)
                        if len(kelompok_dict[i]) == 38:
                            break
            except:
                continue
        iteration += 1
            
    print(f"Sukses {i}")

print("Sukses")
```
